GET FEATURES SELECTION

```python
df_features = df[['Movie_Genre', 'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast', 'Movie_Director']].fillna('')
```

Selected five features to recomended movies. It may vary from one project to another. Like one can add vote counts, budget, language, etc.

```python
x = df_features['Movie_Genre'] + ' ' + df_features['Movie_Keywords']+ ' ' +df_features['Movie_Tagline']+ ' ' +df_features['Movie_Cast'] + ' ' + df_features['Movie_Director']
```

GET FEATURES TEXT CONVERSION TO TOKENS

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x=tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

GET SIMILAR SCORE USONG COSINE SIMILARITY

Cosine_similarity computes the L2 normalized dot product of vector. Euclidean(L2) normalization projects the vector on the unot sphere, and their dot product is then the cosine of the angle between th =e point denoted by the vectors.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

GET MOVIES NAME AS INPUT FROM THE USER AND VALIDATE FOR THE CLOSEST SPELLING

```python
  Favourite_Movie_Name = input("Enter your favourite movie name: ")
```

```python
All_Movies_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recomendation = difflib.get_close_matches(Favourite_Movie_Name, All_Movies_Title_List)
print(Movie_Recomendation)
```

```python
Close_Match = Movie_Recomendation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
print(Index_of_Close_Match_Movie)
```

```python
#getting a list of similar movies
Recomendation_Score = list((enumerate(Similarity_Score[Index_of_Close_Match_Movie])))
print(Recomendation_Score)
```

```python
len(Recomendation_Score)
```

GET ALL MOVIES AORT BASED ON RECOMEDATION SCORE WRT FAVOURITE MOVIE

```python
# sorting the movies based on their similarity score
Sorted_Similar_Movies = sorted(Recomendation_Score, key  = lambda a:a[1], reverse = True)
print(Sorted_Similar_Movies)
```

```python
#print the name of the simialr movies based on the index
print("Top 30 Movies Suggested for you : \n")
i=1
for  movies in Sorted_Similar_Movies:
  index = movies[0]
  title_from_index = df[df.index == index]['Movie_Title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```

TOP 10 MOVIE RECOMMENDATION SYSTEM

```python
Movie_Name = input('Enter your favourite movie name : ')
list_of_all_titles = df['Movie_Title'].tolist()
find_Close_Match = difflib.get_close_matches(Movie_Name,list_of_all_titles)
Close_Match = find_Close_Match[0]
Index_of_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
Recomendation_Score = list(enumerate(Similarity_Score[Index_of_Movie]))
sorted_similar_movies = sorted(Recomendation_Score,key = lambda x:x[1], reverse =True)
print("top 10 movies suggested for you are : \n")
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.Movie_ID == index ]['Movie_Title'].values
  if(i<11):
    print(i, '.', title_from_index)
    i+=1
```
